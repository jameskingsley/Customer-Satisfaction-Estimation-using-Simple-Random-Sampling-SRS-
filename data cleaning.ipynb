```python
#Selecting relevant columns for the project
df_clean = df[['brand', 'categories', 'name', 'reviews.rating', 'reviews.date']]
```

```python
# Converting rating to numeric (float)
df_clean['reviews.rating'] = pd.to_numeric(df_clean['reviews.rating'], errors='coerce')
```

```python
#drop rows with missing values
df_clean = df_clean.dropna(subset=['reviews.rating']).reset_index(drop=True)

print(f"Cleaned dataset shape: {df_clean.shape}")
print(df_clean.head())
```

```python
#cleaning review date
df_clean['reviews.date'] = pd.to_datetime(df_clean['reviews.date'], errors='coerce')
```

```python
#verifying the actions
print(df_clean.info())
print(df_clean['reviews.rating'].describe())
```

```python
#saving cleaned dataset
df_clean.to_csv("cleaned_reviews.csv", index=False)
print("Cleaned dataset saved as 'cleaned_reviews.csv'")
```
